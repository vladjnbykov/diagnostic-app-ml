## Data encoding

```python
# Encoding features
from sklearn.preprocessing import LabelEncoder
```

```python
LE = LabelEncoder()
```

```python
col_encode = data.select_dtypes(include='O').columns
col_encode
```

```python
# removing gender and class from label encoding
col_to_label_encode = ['polyuria', 'polydipsia', 'sudden_weight_loss', 'weakness',
       'polyphagia', 'genital_thrush', 'visual_blurring', 'itching',
       'irritability', 'delayed_healing', 'partial_paresis',
       'muscle_stiffness', 'alopecia', 'obesity']
```

```python
for col in col_to_label_encode:
    data[col] = LE.fit_transform(data[col].astype(str))
```

```python
data.head()
```

```python
data.info()
```

```python
print(LE.classes_)
```

```python
data['gender'].value_counts()
```

```python
data['class'].value_counts()
```

```python
gender_map = {'Female':0, 'Male':1}
class_map = {'Negative':0, 'Positive':1}
```

```python
data['gender'] = data['gender'].map(gender_map)
```

```python
data['class'] = data['class'].map(class_map)
```

```python
data.head()
```

```python
data.describe()
```

# Select ML model

## Train test split

```python
# Select target variable
target_col='class'
```

```python
X = data.drop(target_col, axis = 1)
y = data[target_col]
```

```python
X
```

```python
feature_names = X.columns
feature_names
```

```python
y
```

```python
# Scaling

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
# Splitting dataset

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

## Train and evaluation: training data


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
classifiers = [
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    KNeighborsClassifier(),
    LogisticRegression()
]
```

```python
from sklearn.model_selection import cross_validate
```

```python
for i in classifiers:
    cv_results = cross_validate(i, X_train, y_train, cv = 5, scoring = "accuracy")
    print(i, round(cv_results.get('test_score').mean(), 3))
```

## Grid search: hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'criterion': ['gini', 'entropy'],
             'n_estimators': [50, 100, 200]}
grid = GridSearchCV(RandomForestClassifier(), param_grid, scoring = 'accuracy', cv = 5)
grid.fit(X_train, y_train)
best_score = grid.best_score_
best_params = grid.best_params_
print("Best score:", round(best_score,3))
print("(Hyper)parameters setting for best score:", best_params)
```

```python
best_model = grid.best_estimator_
best_model
```

```python
best_params
```

## Test prediction

```python
best_model.score(X_test, y_test)
```

```python
y_pred = best_model.predict(X_test)
```

### Evaluation of model

```python
from sklearn.metrics import classification_report, confusion_matrix, plot_confusion_matrix, plot_precision_recall_curve, plot_roc_curve
```

```python
target_names = ['Negative(0)', 'Positive(1)']
```

```python
# Classification report
print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
# Confusion matrix
confusion_matrix(y_test, y_pred)
```

```python
plot_confusion_matrix(best_model, X_test, y_test)
```

```python
# ROC curve
# Receiver operating characteristic

import matplotlib.pyplot as plt 

plot_roc_curve(best_model, X_test, y_test)
plt.title('ROC curve for Random Forest model')
plt.show()
```

## Saving the model

## Feature importance, method II

```python
# feature importance
import time
from sklearn.inspection import permutation_importance
start_time = time.time()
result = permutation_importance(
best_model, X_test, y_test, n_repeats=10, n_jobs=2)
elapsed_time = time.time() - start_time
print(f"Elapsed time to compute the importances: "
      f"{elapsed_time:.3f} seconds")

forest_importances = pd.Series(result.importances_mean, index=feature_names)
```

```python
forest_importances
```

```python
forest_importances.sort_values(ascending= False)
```

```python
fig, ax = plt.subplots()
forest_importances.plot.bar(yerr=result.importances_std, ax=ax)
ax.set_title("Feature importances using permutation on full model")
ax.set_ylabel("Mean accuracy decrease")
fig.tight_layout()
plt.show()
```
